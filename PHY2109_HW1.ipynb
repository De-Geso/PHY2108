```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian(x, mu, var):
    return (
        1.0 / (np.sqrt(2.0 * np.pi * var)) * np.exp(-(x - mu)**2 / (2 * var))
    )

x = np.linspace(170,230,1000)
s = 30
peel = gaussian(x,198,s)
pinch = gaussian(x,202,s)


fig, axs = plt.subplots(2)
axs[0].set_title('Probability Distribution')
axs[0].plot(x,peel,label="Peelers")
axs[0].set_xlabel('Height (cm)')
axs[0].vlines(x = 198, ymin = 0, ymax = max(peel), linestyle='--')
axs[0].plot(x,pinch,label="Pinchers")
axs[0].vlines(x = 202, ymin = 0, ymax = max(pinch), linestyle='--', colors = 'tab:orange')

axs[1].set_title(r'$\mathcal{P}$(Party|Height)')
axs[1].plot(x,peel/(peel+pinch))
axs[1].vlines(x = 198, ymin = 0, ymax = 1, linestyle='--')
axs[1].plot(x,pinch/(peel+pinch))
axs[1].vlines(x = 202, ymin = 0, ymax = 1, linestyle='--', colors = 'tab:orange')
axs[0].set_xlabel('Height (cm)')
fig.legend()
fig.tight_layout()
plt.show()
```
